```
lst1 = [1, 2, 3, 4]
lst2 = [3, 4, 5, 6, 7, 8]
s1 = set(lst1)
s2 = set(lst2)
print(s1)
print(s2)
```

```
print(s1 & s2)
print(s1.difference(s2))
print(s2.difference(s1))
print(s1.intersection(s2))
print(s1.union(s2))
```

```
print(s1)
print(s2)
```

```
print(set([1,2,3,4,5]).issuperset(s2))
print(set([2, 3]).issubset(s2))
```
