```
lst = [1, 2, 3, 1, 0]
for i in range(len(lst)):
  print(lst[i])
```

```
# break
for i in range(10):
  print(i)
  if i % 5 == 0 and i > 0:
    break
```

```
#continue
for i in range(10):
  if i % 3 == 0:
    continue
  print(i)
```

```
lst = [1, 2, 4, 6]
i = 0
tmp = 0
while tmp != 4:
  tmp = lst[i]
  print(tmp)
  i += 1
print('i = ', i)
```

```
lst = [1, 2, 3]
i = 0
while i < len(lst):
  print(lst[i])
  if i >= 1:
    break
  i += 1
```
