```
import numpy as np
```

```
a = np.array([1, 2, 3, 4])
b = np.array([3, 4, 5, 6])
print(a)
print(b)
print(a.shape)
print(b.shape)
```

```
print(a * b)
print(np.inner(a, b))
```

```
a = np.array([
  [1, 2, 3],
  [3, 4, 5]
])
b = np.array([
  [3, 4, 5],
  [1, 2, 0],
  [3, 4, 1]
])
print(a)
print(a.shape)
print(b)
print(b.shape)
```

```
c = a.dot(b)
print(c)
print(c.shape)
```

```
a = np.array([
  [1, 2],
  [3, 3]
])
b = np.array([
  [3, 4],
  [1, 9]
])
print(a)
print(a.shape)
print(b)
print(b.shape)
```

```
print(np.tensordot(a, b))
```

```
from numpy import linalg
a = np.array([1, 2, 3])
print(a)
print(linalg.norm(a)) #l2 norm
```

```
a = np.array([1, 2, -3])
print(a)
print(linalg.norm(a, ord = 1)) #l1 norm
```

```
a = np.array([
  [1, 2, 3],
  [3, 4, -1]
])
b = np.array([
  [3, 4, 5],
  [1, 2, 9]
])
print(np.subtract(a,b))
```
