```
def showValue(a):
  if a % 2 == 0:
    print(a, ' la so chan')
  else:
    print(a, ' la so le')
showValue(11)
showValue(110)
```

```
def sumDigit(n):
  res = 0
  while n > 0:
    res += (n % 10)
    n //= 10
  return res
def showValue(n):
  tmp = sumDigit(n)
  if tmp % 2 == 0:
    print('sumdigit of ', n, ' la chan')
  else:
    print('sumdigit of ', n, ' la le')
  
showValue(111)
showValue(121)
```

```
import math
def checkPrime(n):
  can = (int)(math.sqrt(n))
  for i in range(2, can + 1):
    if n % i == 0:
      return False
  return True
def showValue(n = 97):
  tmp = checkPrime(n)
  if tmp == True:
    print(n, 'la so nguyen to')
  else:
    print(n, 'khong la so nguyen to')

showValue()
showValue(12)
```
