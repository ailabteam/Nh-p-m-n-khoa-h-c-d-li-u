```
import numpy as np
np_vector = np.array([1, 2, 3])
print(np_vector)
print(type(np_vector))
print(np_vector.shape)
```

```
np_matrix = np.array([
  [1, 2, 3], 
  [3, 4, 5]
])
print(np_matrix)
print(type(np_matrix))
print(np_matrix.shape)
```

```
np_cube = np.array([
  [[1, 2], [4, 3], [5, 9]],
  [[3, 1], [0, 1], [9, 1]]
])
print(np_cube)
print(type(np_cube))
print(np_cube.shape)
```
