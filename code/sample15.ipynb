```
import numpy as np
```

```
np_vector = np.array([1, 2, 3, 4])
print(np_vector)
#np_vector = np_vector + 3
np_vector = np_vector - 3 #np.array([3, 3, 3, 3])
print(np_vector)
```

```
np_matrix = np.array([
  [1, 2, 3],
  [4, 5, 8]
])
print(np_matrix)
np_matrix = np_matrix * 2
print(np_matrix)
```

```
np_vector = np.arange(0, 20, 3)
print(np_vector)
np_vector *= 3
print(np_vector)
```

```
np_matrix = np.ones((3, 5))
print(np_matrix)
np_matrix *= 2
print(np_matrix)
```

```
a = np.ones((3, 5))
b = np.arange(5)
print(a)
print(b)
```

```
c = a + b
print(c)
```

```
print(c.shape)
```
