```
class Color:
  def __init__(self, color, weight = 0):
    self.color = color
    self.weight = weight
  def display(self):
    print(self.color, self.weight)
  def __lt__(self, other):
    return self.weight > other.weight
```

```
c = Color('Blue', 9)
c.display()
```

```
d = Color('Yellow')
d.display()
```

```
d.weight = 22
d.display()
```

```
e = Color(weight = 33, color = 'Red')
e.display()
```

```
from queue import PriorityQueue
```

```
q = PriorityQueue()
c1 = Color('Red', 10)
c2 = Color('Yellow', 19)
c3 = Color('Blue', 12)
q.put(c1)
q.put(c2)
q.put(c3)

while not q.empty():
  tmp = q.get()
  tmp.display()
```
