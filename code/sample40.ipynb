```
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')

import pandas as pd
path = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx'
df = pd.read_excel(path, sheet_name='Canada by Citizenship', skiprows=range(20), skipfooter=2)

#preprocess
df.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis = 1, inplace=True)
df.rename(columns={'OdName': 'Country', 'AreaName': 'Continent', 'RegName': 'Region'}, inplace=True)
df.columns = list(map(str, df.columns))
df.set_index('Country', inplace=True)
df['Total'] = df.sum(axis = 1)
years = list(map(str, range(1980, 2014)))
```

```
df.head()
```

```
df_dsn = df.loc[['Denmark', 'Norway', 'Sweden'], :]
total_values = sum(df_dsn['Total'])
category_proportions = [(float(value) / total_values) for value in df_dsn['Total']]
for i, value in enumerate(category_proportions):
  print(df_dsn.index.values[i], ': ', str(value))
```

```
width = 40
height = 10
total_num_tiles = width * height
```

```
import numpy as np
```

```
tiles_per_category = [(round(value * total_num_tiles)) for value in category_proportions]
```

```
for i, value in enumerate(tiles_per_category):
  print(df_dsn.index.values[i], ': ', str(value))
```

```
waffle_chart = np.zeros((height, width))
category_index = 0
tile_index = 0
for i in range(width):
  for j in range(height):
    tile_index += 1
    if tile_index > sum(tiles_per_category[0:category_index]):
      category_index += 1
    waffle_chart[j, i] = category_index
```

```
colormap = plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()
```

```
for i in range(width):
  for j in range(height):
    print(waffle_chart[j, i])
  print()
```
