```
import numpy as np
a = np.array([
  [1, 2, 3, 4],
  [3, 4, 5, 6]
])
print(a)
print(a.shape)
```

```
b = a.reshape(4, 2)
print(b)
print(b.shape)
```

```
c = a.reshape(2, 2, 2)
print(c)
print(c.shape)
```

```
a = np.random.random((2, 2, 2))
print(a)
print(a.shape)
```

```
b = a.flatten()
print(b)
print(b.shape)
```

```
a = np.random.random((2, 2))
print(a)
print(a.shape)
```

```
b = np.expand_dims(a, axis = 2)
print(b)
print(b.shape)
```

```
a = np.random.random((1, 2, 2, 1))
print(a)
print(a.shape)
```

```
b = np.squeeze(a, axis = 0)
print(b)
print(b.shape)
```

```
a = np.random.random((2, 3))
print(a)
print(a.shape)
```

```
b = a.T
print(b)
print(b.shape)
```

```
c = np.transpose(a)
print(c)
print(c.shape)
```
